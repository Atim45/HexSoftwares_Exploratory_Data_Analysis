```python
data["Embarked"].fillna(data["Embarked"].mode()[0], inplace=True)
```

```python
zero_cols = [col for col in data.columns if "zero" in col]
data.drop(columns=zero_cols, inplace=True)
```

```python
data.head()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
sns.histplot(data["Age"], kde=True)

plt.title("Age Distribution")
plt.show()
```

##Observation

Most passengers were between 20 and 40 years of age. The distribution appears slightly right-skewed, with a small number of elderly passengers and infants.

```python
plt.figure(figsize=(8,5))
sns.histplot(data["Fare"], kde=True)

plt.title("Fare Distribution")
plt.show()
```

##Observation

Fare distribution is highly positively skewed. Most passengers paid low fares, while a few passengers paid exceptionally high fares, creating outliers.

```python
sns.countplot(x="Pclass", data=data)

plt.title("Passenger Class Distribution")
plt.show()
```

##Observation

Third-class passengers represent the largest group in the dataset, indicating that lower-class travel was the most common.

```python
sns.countplot(x="2urvived", data=data)

plt.title("Survival Count")
plt.show()
```

##Observation

Only about 26% of passengers survived, while nearly 74% did not survive the disaster.

```python
sns.countplot(x="Sex",hue="2urvived",data=data)

plt.title("Gender vs Survival")
plt.show()
```

##Observation

Female passengers had a significantly higher survival rate compared to male passengers, suggesting gender played an important role in rescue priority.

```python
sns.countplot( x="Pclass",hue="2urvived",data=data)

plt.title("Pclass vs Survival")
plt.show()
```

##Observation

First-class passengers had a noticeably higher survival rate than second- and third-class passengers.

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

##Observation

Fare shows a positive relationship with survival, while passenger class shows a negative relationship with survival. Age has a relatively weak correlation with survival.

```python
sns.boxplot(x="2urvived",y="Age",data=data)

plt.show()
```

##Observation

The age distributions of survivors and non-survivors overlap considerably, indicating that age alone was not a strong determinant of survival.

```python
sns.boxplot(x="2urvived",y="Fare",data=data)

plt.show()
```

##Observation

Survivors generally paid higher fares than non-survivors, suggesting passengers with higher ticket classes had better survival chances.

```python
data["FamilySize"] = data["sibsp"] + data["Parch"] + 1
```

```python
sns.countplot(x="FamilySize", data=data)
plt.title("Family Size Distribution")
plt.show()
```

## Observation

Most passengers traveled either alone or with very small families.
Large family groups were relatively uncommon.
This suggests that solo travel was more frequent among Titanic passengers.
